### Djikstra Algorithm

```python
import networkx as nx

def dijkstra_shortest_path(graph, start_id, end_id):
    """
    Compute the shortest path between two nodes using Dijkstra's algorithm.
    
    Parameters:
        - graph: NetworkX graph object
        - start_id: ID of the starting node
        - end_id: ID of the ending node
        
    Returns:
        - shortest_path: List of nodes representing the shortest path
        - shortest_distance: Shortest distance between the start and end nodes
    """
    shortest_path = nx.shortest_path(graph, source=start_id, target=end_id, weight='length')
    shortest_distance = nx.shortest_path_length(graph, source=start_id, target=end_id, weight='length')
    return shortest_path, shortest_distance

start_node_id = '0'  # Example starting node ID
end_node_id = '20427'  # Example ending node ID

if start_node_id not in graph.nodes:
    print("Error: Starting node ID is not in the graph.")
elif end_node_id not in graph.nodes:
    print("Error: Ending node ID is not in the graph.")
else:
    shortest_path_dijkstra, shortest_distance_dijkstra = dijkstra_shortest_path(graph, start_node_id, end_node_id)
    print("Shortest path using Dijkstra's Algorithm:", shortest_path_dijkstra)
    print("Shortest distance using Dijkstra's Algorithm:", shortest_distance_dijkstra)
```

### Astar Algorithm

```python
import networkx as nx

def astar_shortest_path(graph, start_id, end_id):
    """
    Compute the shortest path between two nodes using A* search algorithm.
    
    Parameters:
        - graph: NetworkX graph object
        - start_id: ID of the starting node
        - end_id: ID of the ending node
        
    Returns:
        - shortest_path: List of nodes representing the shortest path
        - shortest_distance: Shortest distance between the start and end nodes
    """
    shortest_path = nx.astar_path(graph, source=start_id, target=end_id, heuristic=None, weight='length')
    shortest_distance = nx.astar_path_length(graph, source=start_id, target=end_id, heuristic=None, weight='length')
    return shortest_path, shortest_distance

# Example usage:
start_node_id = '0'  # Example starting node ID
end_node_id = '20427'  # Example ending node ID

if start_node_id not in graph.nodes or end_node_id not in graph.nodes:
    print("Error: Starting or ending node ID is not in the graph.")
else:
    shortest_path_astar, shortest_distance_astar = astar_shortest_path(graph, start_node_id, end_node_id)
    print("Shortest path using A* search Algorithm:", shortest_path_astar)
    print("Shortest distance using A* search Algorithm:", shortest_distance_astar)
```

### Visualizing the shortest path

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_shortest_path(graph, shortest_path):
    """
    Visualize the graph with the shortest path highlighted.
    
    Parameters:
        - graph: NetworkX graph object
        - shortest_path: List of nodes representing the shortest path
    """
    # Plot the graph
    plt.figure(figsize=(10, 10))
    if 'lon' in graph.nodes[next(iter(graph.nodes))]:  # Check if 'lon' attribute is available
        node_positions = {node: (graph.nodes[node]['lon'], graph.nodes[node]['lat']) for node in graph.nodes}
        nx.draw(graph, pos=node_positions, node_size=10, alpha=0.6)
    else:
        nx.draw(graph, node_size=10, alpha=0.6)
    
    # Draw the shortest path with a different color
    nx.draw_networkx_nodes(graph, pos=node_positions, nodelist=shortest_path, node_color='r', node_size=20)
    nx.draw_networkx_edges(graph, pos=node_positions, edgelist=[(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)], edge_color='r', width=2)
    
    plt.title("Shortest Path on the Map")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()

# Example usage:
start_node_id = '0'  # Example starting node ID
end_node_id = '20427'  # Example ending node ID

if start_node_id not in graph.nodes or end_node_id not in graph.nodes:
    print("Error: Starting or ending node ID is not in the graph.")
else:
    shortest_path_dijkstra, shortest_distance_dijkstra = dijkstra_shortest_path(graph, start_node_id, end_node_id)
    print("Shortest path using Dijkstra's Algorithm:", shortest_path_dijkstra)
    print("Shortest distance using Dijkstra's Algorithm:", shortest_distance_dijkstra)
    visualize_shortest_path(graph, shortest_path_dijkstra)
```
